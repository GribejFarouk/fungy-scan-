```python
from ultralytics import YOLO
from IPython.display import Image

# Load the trained YOLOv8 model
model = YOLO('yolov8n.pt')  # Replace with the path to your trained model

# Test on a new image
results = model.predict('mnayka2.jpg', save=True)

# Check if any infection is detected
infection_detected = False
for result in results:
    for box in result.boxes:
        class_id = int(box.cls)  # Class ID of the detected object
        confidence = float(box.conf)  # Confidence score
        if class_id == 1 and confidence > 0.5:  # Assuming class_id 1 represents 'infected'
            infection_detected = True
            break

# Display the result
if infection_detected:
    print("The plant is infected.")
else:
    print("The plant is healthy.")

# Display the image with bounding boxes
Image('mnayka2.jpg')
```
